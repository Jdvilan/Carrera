```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Leer el archivo Excel
df = pd.read_excel('Alumnos.xlsx', sheet_name='in')
np.random.seed(42)

# a) Nombre completo
df['Nombre Completo'] = df['Nombre'] + ' ' + df['Apellido']

# b) Edad entre 18 y 25 años
n_alumnos = len(df)
df['Edad'] = np.random.randint(18, 26, size=n_alumnos)

# c) Calificación con distribución normal (media=7.5, desviación=1.0)
df['Calificación'] = np.random.normal(loc=7.5, scale=1.0, size=n_alumnos)

# d) Asistencia entre 70% y 100% (uniforme)
df['Porcentaje de Asistencia'] = np.random.uniform(70, 100, size=n_alumnos)

# Seleccionar y formatear las columnas finales
resultado = df[['Nombre Completo', 'Edad', 'Calificación', 'Porcentaje de Asistencia']]

# Redondear calificación y asistencia a 2 decimales
resultado['Calificación'] = resultado['Calificación'].round(2)
resultado['Porcentaje de Asistencia'] = resultado['Porcentaje de Asistencia'].round(2)
```

```python
print(resultado)
```

```python
# Extraer datos de edades
edades = resultado['Edad']

# Configurar el histograma
plt.figure(figsize=(10, 6))
plt.hist(edades,
         bins=range(18, 26),
         edgecolor='black',
         color='skyblue',
         align='left')

# Personalización
plt.title('Distribución de Edades de los Alumnos', fontsize=14)
plt.xlabel('Edad', fontsize=12)
plt.ylabel('Cantidad de Alumnos', fontsize=12)
plt.xticks(range(18, 26))
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar
plt.show()
```

```python
# Configurar tamaño del gráfico
plt.figure(figsize=(10, 6))

# Crear el gráfico de barras
plt.bar(resultado['Nombre Completo'],
        resultado['Porcentaje de Asistencia'],
        color='teal')

# Personalización
plt.title('Porcentaje de Asistencia por Alumno', fontsize=14, pad=20)
plt.xlabel('Alumnos', fontsize=12)
plt.ylabel('Asistencia (%)', fontsize=12)

# Rotar nombres verticalmente y ajustar límites
plt.xticks(rotation=90, fontsize=8)
plt.ylim(50, 100)  # Rango desde 50% hasta 100%

# Añadir cuadrícula y ajustar layout
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```

```python
# Configurar el tamaño del gráfico
plt.figure(figsize=(10, 6))

# Crear el scatter plot
plt.scatter(x = resultado['Edad'],
            y = resultado['Porcentaje de Asistencia'],
            color = 'royalblue',
            alpha = 0.7,
            edgecolor = 'black')

# Personalización del gráfico
plt.title('Relación entre Edad y Asistencia', fontsize=14, pad=15)
plt.xlabel('Edad (años)', fontsize=12)
plt.ylabel('Porcentaje de Asistencia (%)', fontsize=12)

# Configurar ejes y retícula
plt.xticks(range(18, 26))
plt.ylim(60, 101)
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
# Configurar tamaño del gráfico
plt.figure(figsize=(10, 6))

# Crear el gráfico de barras
plt.bar(resultado['Nombre Completo'],
        resultado['Calificación'],
        color='coral')

# Personalización
plt.title('Calificaciones de los Alumnos', fontsize=14, pad=20)
plt.xlabel('Alumnos', fontsize=12)
plt.ylabel('Calificación', fontsize=12)

# Rotar nombres y ajustar ejes
plt.xticks(rotation=90, fontsize=8)
plt.ylim(5, 10)

# Añadir elementos de referencia
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Línea de referencia para la media
plt.axhline(y=7.5,
            color='dimgray',
            linestyle=':',
            linewidth=1.5,
            label='Media teórica (7.5)')

plt.legend()
plt.show()
```

```python
# Calcular estadísticos y filtrar
estadisticos = resultado.describe().loc[['mean', 'std', 'min', 'max']].round(2)

# Renombrar las filas en español
estadisticos = estadisticos.rename(index={
    'mean': 'Media',
    'std': 'Desviación Estándar',
    'min': 'Mínimo',
    'max': 'Máximo'
})

print(estadisticos)
```
