**1.-Determine el ciclo o periodo de vida de los siguientes generadores congruenciales**

```python
from prettytable import PrettyTable

def tabla(iteraciones, x0, a, b, m):
    #Iniciamos la tabla con los encabezados "Iteración" y "xn"
    t = PrettyTable(["Iteración", "xn"])
    #Configuramos el valor inicial para xn
    xn = x0
    #Definimos la formula
    funcion = lambda x: (a * x + b) % m
    #Iteramos hasta que 'iteraciones' sea igual al número de iteraciones solicitadas
    for i in range(iteraciones):
        xn1 = funcion(xn) #Calculamos el siguiente valor de xn
        t.add_row([i + 1, xn]) #Añadimos los valores de las iteraciones y los xn a la tabla
        xn = xn1 #Actualizamos el valor de xn para la siguiente iteración

    return t
```

```python
from prettytable import PrettyTable

def tabla(iteracion, x0, a, b, m, x1):
    t = PrettyTable(["Iteración", "xn"])
    xn = x1
    xj = x0
    funcion = lambda x0, x1: (a * x1 + b * x0) % m
    t.add_row([0, xj])
     #Creamos un diccionario para guardar los valores de xn y su numero de iteración
    valorv = {xj: 0}

    for i in range(1, iteracion + 1):
        xn1 = funcion(xj, xn)
        #Verificamos que el valor xn1 no se repita
        if xn1 in valorv:
            print(f"El valor {xn1} se repite en la iteración {i}")
            break

        t.add_row([i, xn1])
        valorv[xn1] = i
        xj = xn
        xn = xn1

    return t

tabla(200, 21, 21, 15, 64, 43)
```

```python
datos = [0.8797, 0.3884, 0.6289, 0.8750, 0.5999, 0.8589, 0.9996, 0.2415, 0.3808, 0.9606, 0.9848, 0.3469, 0.7977, 0.5844, 0.8147, 0.6431, 0.7387, 0.5613, 0.0318,
         0.7401, 0.4557, 0.1592, 0.8536, 0.8846, 0.3410, 0.1492, 0.8681, 0.5291, 0.3188, 0.5992, 0.9170, 0.2204, 0.5991, 0.5461, 0.5739, 0.3254, 0.0856, 0.2258,
         0.4603, 0.5027, 0.8376, 0.6235, 0.3681, 0.2088, 0.1525, 0.2006, 0.4720, 0.4272, 0.6360, 0.0954]
```

$H_0$: No hay diferencia entre frecuencias observadas y frecuencias esperadas.

$H_1$: Hay una diferencia entre las frecuencias

Para un nivel de significancia de 0.05.

El estadistico de prueba sigue la distribución
$$\chi^2_{n-1}=\sum\frac{(f_0-f_e)^2}{f_e}$$

```python
from tabulate import tabulate

#Definimos los intervalos
intervalos = [(0, 0.2), (0.2, 0.4), (0.4, 0.6), (0.6, 0.8), (0.8, 1.0)]

#Iniciamos el contador para cada intervalo
conteo_intervalos = [0] * len(intervalos)

#Contamos cuantos datos caen en cada intervalo
for dato in datos:
    for i, (limite_inferior, limite_superior) in enumerate(intervalos):
        if limite_inferior <= dato < limite_superior:
            conteo_intervalos[i] += 1
            break

#Creamos la tabla con los resultados
tabla_resultados = []
for i, (lim_inf, lim_sup) in enumerate(intervalos):
    cantidad_datos = conteo_intervalos[i]
    resultado_operacion = ((cantidad_datos - 10) ** 2) / 10 if cantidad_datos != 0 else 0
    tabla_resultados.append([f'[{lim_inf}, {lim_sup})', cantidad_datos, 10, resultado_operacion])

print(tabulate(tabla_resultados, headers=['Clases', 'f_0', 'f_e', '(f_0-f_e)^2/f_e'], tablefmt='grid'))
```

El valor critico esta dado por
$$\chi_{n-1}^2$$
donde $n$ es el número de clases.

Por lo tanto, buscamos en la tabla de valores para la distribución $\chi_n^2$, el valor que tenga 4 grados de libertad y un área de cola derecha de 0.05, obteniendo como resultado 9.488.

Así, la regla de decisión nos dice que debemos rechazar $H_0$ si el valor calculado de $\chi_{n-1}^2>9.488$.

```python
def chi(tabla):
    suma = 0
    for fila in tabla:
        suma += fila[3]
    return suma

total_suma = chi(tabla_resultados)

if total_suma > 9.488:
  print('Rechazar H_0')
else:
  print(f'{total_suma} < 9.488')
  print('No rechazar H_0')
```

Por lo tanto, podemos decir que no hay diferencia entre frecuencias observadas y frecuencias esperadas.

```python
def tabla(iteraciones, x0, a, b, m):
    valores_xn = []
    xn = x0
    funcion = lambda x: (a * x + b) % m
    for _ in range(iteraciones):
        xn1 = funcion(xn)
        valores_xn.append(xn)
        xn = xn1

    return valores_xn

valores_xn = tabla(30, 21, 21, 15, 31)
```
