**Distribución triangular**

Supongamos los valores $a=2$, $b=3$ y $c=6$ para poder graficar la distribución.

```python
import numpy as np
import random as rnd
import matplotlib.pyplot as plt
```

```python
def triangular(x, a, b, c):
    if x < a or x > c:
        return 0
    elif x < b:
        return 2 * (x - a) / ((b - a) * (c - a))
    else:
        return 2 * (c - x) / ((c - a) * (c - b))

# Generar puntos para la curva
x = np.linspace(2, 6, 1000)
y = [triangular(val, 2, 3, 6) for val in x]

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución triangular', color='darkviolet')

# Agregar una línea vertical en x=3 solo en el intervalo (0, 0.5)
y_interval = np.linspace(0, 0.5, 100)
x_interval = np.full_like(y_interval, 3)
plt.plot(x_interval, y_interval, color='deeppink', linestyle='--', label='x=3 en (0, 0.5)')

# Agregar una línea horizontal en y=0.5 solo en el intervalo (0, 3)
x_interval = np.linspace(0, 3, 100)
y_interval = np.full_like(x_interval, 0.5)
plt.plot(x_interval, y_interval, color='deeppink', linestyle='--', label='y=0.5')

# Títulos y etiquetas
plt.title('Distribución Triangular')
plt.xlabel('x')
plt.ylabel('f(x)')

# Mostrar la gráfica
plt.grid(True)
plt.legend()
plt.show()
```

El método de composición nos da una serie de pasos

**Paso 1)** Dividir la distribución en subareas.

Lo resolvemos para un caso general, las subareas del triangulo estan divididas por la linea roja punteada de la grafica.

El área 1 esta dada por:

$$A_1=\frac{bh}{2}=\frac{(b-a)\left(\frac{2}{c-a}\right)}{2}=\frac{b-a}{c-a}$$

El área 2:

$$A_2=\frac{bh}{2}=\frac{(c-b)\left(\frac{2}{c-a}\right)}{2}=\frac{c-b}{c-a}$$

**Paso 2)** Definir una distribución para cada subarea.

Cada subarea debe cumplir

$$\int_{-∞}^∞f_i(x)dx=1,\hspace{0.5cm}∀i$$

Entonces,

$$A_1=1=\frac{bh_1}{2}⟹h_1=\frac{2}{b}=\frac{2}{b-a}$$

Esta es la altura para $A_1$, con esta altura tenemos los puntos

$$v=\left(a,0\right),\hspace{1cm}w=\left(b,\frac{2}{b-a}\right)$$

Sustituyendo los puntos en la formula punto pendiente encontramos $f_1(x)$:

$$y-y_1=\frac{y_2-y_1}{x_2-x_1}(x-x_1)$$
$$\Longrightarrow y-0=\frac{\frac{2}{b-a}-0}{b-a}(x-a)$$
$$\Longrightarrow y=\frac{2(x-a)}{(b-a)^2}$$

Analogamente para $f_2(x)$:

$$A_2=1=\frac{bh_2}{2}⟹h_2=\frac{2}{b}=\frac{2}{c-b}$$
$$v=\left(b,\frac{2}{c-b}\right),\hspace{1cm}w=\left(c,0\right)$$
$$y-\frac{2}{c-b}=\frac{0-\frac{2}{c-b}}{c-b}(x-b)$$
$$\Longrightarrow y=-\frac{2(x-b)}{(c-b)^2}+\frac{2}{c-b}=-\frac{2(x-c)}{(c-b)^2}$$

Así, la función de densidad de probabilidad es:

$$
f(x) = \left\{
            \begin{array}{ll}
	      \frac{2(x-a)}{(b-a)^2} & \mathrm{si\ } a<x\leq b \\
	      -\frac{2(x-c)}{(c-b)^2} & \mathrm{si\ } b<x\leq c
            \end{array}
	      \right.\
$$

**Paso 3)** La función de distribución de probabilidad original se expresa como:

$$f(x)=A_1f_1(x)+A_2f_2(x)=\frac{b-a}{c-a}\left(\frac{2(x-a)}{(b-a)^2}\right)+\frac{c-b}{c-a}\left(-\frac{2(x-c)}{(c-b)^2}\right)=\frac{2(x-a)}{(c-a)(b-a)}-\frac{2(x-c)}{(c-a)(c-b)}$$

**Paso 4)** Obtener la distribución acumulada de las áreas.

La función de distribución acumulativa de una variable aleatoria continua $X$ es la probabilidad de que $X$ tome un valor menor o igual a algun $x$ especifico. Esto es:

$$F(x)=\int_{-∞}^x f(t)dt$$

Para $f_1(x)$:

$$F_1(x)=\int_a^x\frac{2(t-a)}{(b-a)^2}dt=\frac{2}{(b-a)^2}\int_a^x (t-a)dt=\frac{2}{(b-a)^2}\left(\frac{t^2}{2}-at\right)\Bigg|_a^x=\frac{2}{(b-a)^2}\left(\frac{x^2}{2}-ax-\frac{a^2}{2}+a^2\right)=\frac{(x-a)^2}{(b-a)^2}$$

Para $f_2(x)$:

$$F_2(x)=\int_b^x -\frac{2(t-c)}{(c-b)^2}dt=-\frac{2}{(c-b)^2}\int_b^x(t-c)dt=-\frac{2}{(c-b)^2}\left(\frac{t^2}{2}-ct\right)\Bigg|_b^x=-\frac{2}{(c-b)^2}\left(\frac{x^2}{2}-cx-\frac{b^2}{2}+bc\right)$$

$$\Longrightarrow F_2(x)=\frac{-x^2+2cx+b^2-2bc+c^2-c^2}{(c-b)^2}=\frac{(c-b)^2-(x-c)^2}{(c-b)^2}=1-\frac{(x-c)^2}{(c-b)^2}$$


Así, la función de distribución acumulativa es:

$$
F(x) = \left\{
            \begin{array}{ll}
	      \frac{(x-a)^2}{(b-a)^2} & \mathrm{si\ } a<x\leq b \\
	      1-\frac{(x-c)^2}{(c-b)^2} & \mathrm{si\ } b<x\leq c
            \end{array}
	      \right.\
$$

```python
# Definir la función de distribución acumulada (CDF) de la distribución triangular
def triangular_F(x, a, b, c):
    if x < a:
        return 0
    elif a <= x < b:
        return ((x - a)**2) / ((b - a)**2)
    elif b <= x < c:
        return 1 - ((x - c)**2) / ((c - b)**2)
    else:
        return 1

# Crear un vector de valores x
x_values = np.linspace(2 - 1, 6 + 1, 1000)
# Calcular los valores de la CDF para cada x
cdf_values = [triangular_F(x, 2, 3, 6) for x in x_values]

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(x_values, cdf_values, label=f'Triangular CDF(a={2}, b={3}, c={6})', color='blue')
plt.title('Función de Distribución Acumulada (CDF) de una Distribución Triangular')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.legend()
plt.grid(True)
plt.show()
```

**Paso 5)** Aplicamos el método de la transformada inversa.

Para $F_1(x)$:

$$\frac{(x-a)^2}{(b-a)^2}=R_2⟹(x-a)^2=R_2(b-a)^2⟹x=a+(b-a)\sqrt{R_2}$$

Para $F_2(x)$:

$$1-\frac{(x-c)^2}{(c-b)^2}=R_2⟹\frac{(x-c)^2}{(c-b)^2}=1-R_2⟹x=c+(c-b)\sqrt{1-R_2}$$

Finalmente, generamos una muestra que siga la distribución:

$$
F^{-1}(U) = \left\{
            \begin{array}{ll}
	      a+(b-a)\sqrt{R_2} & \mathrm{si\ } 0<U\leq \frac{b-a}{c-a} \\
	      c+(c-b)\sqrt{1-R_2} & \mathrm{si\ } \frac{b-a}{c-a}<U\leq 1
            \end{array}
	      \right.\
$$

```python
def triangular(a, b, c, n): # Definir la función y generar n repeticiones
    lista = []
    for i in range(n):
        U1 = rnd.random()
        U2 = rnd.random()
        y = (b-a)/(c-a)
        if 0 <= U1 <= y:
            X = a + (b - a)*U2**(1/2)
        else:
            X = c - (c - b)*(1 - U2)**(1/2)
        lista.append(X)
    return lista

print(triangular(1, 3, 5, 10))
```

```python
G = triangular(1, 3, 5, 10000) # Obtención de media, varianza y mediana
media = np.mean(G)
varianza = np.var(G)
mediana = np.median(G)
print(f'Mediana: {mediana}')
print(f'Media: {media}')
print(f'Varianza: {varianza}')
```

```python
# Generación del grafico de la función de distribución
plt.hist(G, bins=30, density=True, alpha=0.6, color='b')
plt.axvline(media, color='r', linestyle='--', label=f'Media: {media:.2f}') # Media
plt.axvline(mediana, color='g', linestyle='--', label=f'Mediana: {mediana:.2f}') # Mediana

plt.xlabel('Valor')
plt.ylabel('Densidad de probabilidad')
plt.title('Histograma de variables aleatorias exponenciales')
plt.legend()

plt.tight_layout()

plt.tight_layout()
plt.show()
```

```python
# Generación de función acumulativa
def cdf(G):
    n = len(G)
    x = np.sort(G)
    y = np.arange(0, n)/n
    return x,y

X, Y = cdf(G)

plt.step(X, Y)
plt.axvline(media, color='r', linestyle='--', label=f'Media: {media:.2f}') # Media
plt.axvline(mediana, color='g', linestyle='--', label=f'Mediana: {mediana:.2f}') # Mediana

plt.xlabel('Valor')
plt.ylabel('Densidad de probabilidad')
plt.title('Función acumulativa')
plt.legend()

plt.tight_layout()

plt.tight_layout()
plt.show()
```
