*Cadenas de Markov absorbentes*

Un estado $i$ de una cadena de Markov se dice absorbente si es imposible abandonarlo ($p_{ii}=1$).

Una cadena de Markov se dice absorbente si posee al menos un estado absorbente y desde cada estado es posible llegar al estado absorbente (no necesariamente en un paso).

Se puede ver que en una cadena absorbente, todos los estados que no son absorbentes son transitorios.

```python
# PASO 5: Genere números aleatorios.

def dado():
    return rnd.randint(1, 6)

# Generar y mostrar números
for i in range(1):
    print(dado())
```

```python
# PASO 6: Simule una partida del juego.

def juego():
    casilla = 1  # Posición inicial
    l = []  # Lista para almacenar las posiciones
    tiradas = 0  # Contador de las tiradas

    for i in range(1000):
        l.append(casilla)
        tiradas += 1  # Incrementa el contador cada vez que se tira el dado
        casilla += dado()

        if casilla >= 20:
            casilla = 20
            l.append(casilla)
            break
        elif casilla == 3:
            casilla += 8  # Escalera
        elif casilla == 13:
            casilla -= 9  # Serpiente
        elif casilla == 15:
            casilla += 4  # Escalera
        elif casilla == 17:
            casilla -= 7  # Serpiente

    return l, tiradas

# Llamada a la función y mostrar el resultado
lista_posiciones, total_tiradas = juego()
print(f"Posiciones: {lista_posiciones}")
print(f"Total de tiradas: {total_tiradas}")
```

```python
# PASO 7. Obtenga la gráfica de estabilización que indica que el tamaño de muestra utilizado es suficiente para
        # garantizar la convergencia del resultado.

def juego():
    casilla = 1  # Posición inicial
    tiradas = 0  # Contador de las tiradas

    for i in range(1000):
        tiradas += 1  # Incrementa el contador cada vez que se tira el dado
        casilla += dado()

        if casilla >= 20:
            casilla = 20
            break
        elif casilla == 3:
            casilla += 8  # Escalera
        elif casilla == 13:
            casilla -= 9  # Serpiente
        elif casilla == 15:
            casilla += 4  # Escalera
        elif casilla == 17:
            casilla -= 7  # Serpiente

    return tiradas  # Solo devuelve el número de tiradas

def modelo(n):
    total_tiradas = 0
    lista = []

    for i in range(n):
        tiradas = juego()  # Obtiene solo el número de tiradas
        lista.append(tiradas)
        total_tiradas += tiradas  # Suma el número de tiradas

    promedio = total_tiradas / n  # Calcula el promedio
    return promedio, lista

# Llamada a la función y mostrar el promedio
n = 1500
promedio, lista = modelo(n)

# Calcular la media acumulada
media_acumulada = np.cumsum(lista) / np.arange(1, n + 1)

# Graficar la convergencia
plt.figure(figsize=(10, 6))
plt.plot(media_acumulada, label='Tiradas', color='b')  # Media acumulada
plt.axhline(promedio, color='r', linestyle='--', label=f'Media: {promedio:.4f}')  # Línea horizontal en la media
plt.title('Tiradas promedio para completar el juego')
plt.xlabel('Número de simulaciones')
plt.ylabel('Media acumulada de tiradas')
plt.legend()  # Añade leyenda
plt.grid(True)  # Activa la cuadrícula
plt.show()
```

```python
# PASO 8. Replique el modelo.

def simulacion(n):
    a = []  # Lista para almacenar las medias de cada corrida
    for i in range(n):
        promedio, _ = modelo(1500)  # Realiza una corrida de 1500 iteraciones
        a.append(promedio)  # Guarda la media de las estimaciones de esa corrida
    return a  # Devuelve la lista de medias

# Corre el modelo con 100 iteraciones (corridas de 1500 simulaciones cada una)
Y = simulacion(500)

# Genera un histograma de los valores obtenidos en el modelo
plt.hist(Y, bins=15, edgecolor='black')  # Grafica la distribución de las medias
plt.title('Tiradas promedio para completar el juego')
plt.xlabel('Número de tiradas promedio')
plt.ylabel('Frecuencia')
plt.grid(axis='y', alpha=0.75)  # Activa la cuadrícula
plt.show()
```

```python
# PASO 9. Calcule la media del área bajo la curva de la función y su desviación estándar.

media = np.mean(Y)
desviacion = np.std(Y)
print("La media es:", media)
print("La desviación estandar es:", desviacion)
```

```python
# PASO 10. Halle el intervalo de confianza del valor de la integral con un nivel de confianza del 95%

alpha = 1 - 0.95  # Nivel de significancia (5%)
z_alpha_2 = norm.ppf(1 - alpha / 2)  # Valor crítico para un intervalo de confianza del 95%
l_inferior = media - z_alpha_2 * (desviacion / math.sqrt(100))  # Límite inferior del intervalo de confianza
l_superior = media + z_alpha_2 * (desviacion / math.sqrt(100))  # Límite superior del intervalo de confianza
print("Entonces: ", l_inferior, "< \u03BC <", l_superior)  # Imprime el intervalo de confianza
```
