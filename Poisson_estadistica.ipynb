1.- Un alumno de sexto semestre de ESFM no tiene alguna clase de las materias de su horario en promedio 3 veces al mes.

a) Encuentre la Probabilidad de que no tenga 4 clases de su horario el siguiente mes. (interprete)

b) ¿Cuál es la E[x] y la Var[x] de este evento? (interprete)

***Solución analítica***

**Datos:**

Media de clases que no tiene al mes $λ=3$

**Procedimiento:**

Sea la variable aleatoria X el número de veces que no tiene clase en un mes.
$$X∼P(λ=3)$$

**Solución:**

a) $$P(X=4)=\frac{e^{-λ}λ^x}{x!}=\frac{e^{-3}3^4}{4!}=0.1680$$

Hay aproximadamente un 16.8% de probabilidad de que el alumno no tenga clases exactamente 4 veces el siguiente mes.

```python
from scipy.stats import poisson

# Parámetro de Poisson
lmbda = 3

# a) Probabilidad de que no tenga clase 4 veces
prob_4 = poisson.pmf(4, lmbda)

print(f"Probabilidad de que no tenga clases 4 veces en el mes: {prob_4:.4f}")
```

b) $$E[X]=λ=3$$

En promedio, el alumno no tendrá clases 3 veces al mes.

$$Var[X]=λ=3$$

La dispersión del número de clases que pierde varía también alrededor de 3.

```python
import matplotlib.pyplot as plt
import numpy as np

# b) Esperanza y varianza
esperanza = poisson.mean(mu=lmbda)
varianza = poisson.var(mu=lmbda)

# Imprimir resultados
print(f"Esperanza E[X] = {esperanza}")
print(f"Varianza Var[X] = {varianza}")
```

```python
# c) Gráfica de la distribución de Poisson
x = np.arange(0, 11)  # Valores de X (0 a 10 veces sin clase)
y = poisson.pmf(x, lmbda)

plt.figure(figsize=(8, 5))
plt.bar(x, y, color="skyblue", edgecolor="black")
plt.axvline(esperanza, color="red", linestyle="--", label=f"E[X] = {esperanza}")
plt.title("Distribución Poisson (λ = 3)\nNúmero de veces sin clase al mes")
plt.xlabel("Número de veces sin clase")
plt.ylabel("Probabilidad")
plt.xticks(x)
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

2.- El Profesor Paul recibe en promedio 10 correos al día.

a) ¿Cuál es la probabilidad de que reciba **a lo menos** 12 correos en un día.

b) ¿Cuál es la probabilidad de que reciba **menos de** 8 correos en un día.

***Solución analítica***

**Datos:**

Media de clases que no tiene al mes $λ=10$

**Procedimiento:**

Sea la variable aleatoria X el número de correos que recibe en un día.
$$X∼P(λ=10)$$

**Solución:**

a) $$P(X\geq 12)=1-P(X≤11)=1-\sum_{x=0}^{11}\frac{e^{-10}10^x}{x!}=1-0.6968=0.3032$$

Hay aproximadamente un 30.32% de probabilidad de que el profesor Paul reciba 12 o más correos en un día.

b) $$P(X<8)=P(X≤7)=\sum_{x=0}^{7}\frac{e^{-10}10^x}{x!}=0.2202$$

Hay aproximadamente un 22.02% de probabilidad de que reciba menos de 8 correos.

```python
# Parámetro de la distribución
lmbda = 10

# a) P(X >= 12) = 1 - P(X <= 11)
prob_a = 1 - poisson.cdf(11, lmbda)

# b) P(X < 8) = P(X <= 7)
prob_b = poisson.cdf(7, lmbda)

# Imprimir resultados
print(f"a) Probabilidad de que reciba al menos 12 correos: {prob_a:.4f}")
print(f"b) Probabilidad de que reciba menos de 8 correos: {prob_b:.4f}")
```

3.- Omar recibe mensajes de whatsapp a una tasa de 5 mensajes por minuto.

a) ¿Cuál es la probabilidad de que Omar reciba 7 mensajes de whatsapp en un minuto?

b) ¿Cuál es la E[X] y la Var[X] ? (interprete)

***Solución analítica***

**Datos:**

Media de clases que no tiene al mes $λ=5$

**Procedimiento:**

Sea la variable aleatoria X el número de mensajes que Omar recibe en un minuto.
$$X∼P(λ=5)$$

**Solución:**

a) $$P(X=7)=\frac{e^{-5}5^7}{7!}=0.1044$$

Hay un 10.44% de probabilidad de que Omar reciba exactamente 7 mensajes en un minuto.

b) $$E[X]=λ=5$$

$$Var[X]=λ=5$$

En promedio, Omar recibe 5 mensajes por minuto, y la variabilidad también es de 5 mensajes.

```python
# Parámetro de la distribución
lmbda = 5

# a) P(X = 7)
prob_7 = poisson.pmf(7, lmbda)

# b) Esperanza y Varianza
esperanza = poisson.mean(mu=lmbda)
varianza = poisson.var(mu=lmbda)

# Mostrar resultados
print(f"a) Probabilidad de que Omar reciba 7 mensajes en un minuto: {prob_7:.4f}")
print(f"b) Esperanza E[X] = {esperanza}")
print(f"   Varianza Var[X] = {varianza}")
```

```python
# Valores posibles de X
x = np.arange(0, 15)
y = poisson.pmf(x, lmbda)

# Probabilidad puntual en X = 7
x_destacado = 7
y_destacado = poisson.pmf(x_destacado, lmbda)

# Crear la gráfica
plt.figure(figsize=(8, 5))
plt.bar(x, y, color='skyblue', edgecolor='black', label="Distribución de Poisson")
plt.bar(x_destacado, y_destacado, color='orange', edgecolor='black', label="P(X = 7)")

# Línea de la media
plt.axvline(lmbda, color='red', linestyle='--', label=f"E[X] = {lmbda}")

# Títulos y etiquetas
plt.title("Distribución de Poisson (λ = 5)\nMensajes de WhatsApp por minuto")
plt.xlabel("Número de mensajes en un minuto")
plt.ylabel("Probabilidad")
plt.xticks(x)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
