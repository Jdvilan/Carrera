```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Función periodica

def f(x):
    if 0 <= x <= 2 * np.pi:
        return 0.5 * (np.pi - x)
    else:
        return f(x + 2 * np.pi if x < 0 else x - 2 * np.pi)

# Sumas parciales
def sum_sin(x, n):
    return sum(np.sin(k * x) / k for k in range(1, n + 1))

# Definir el rango de valores de x
x = np.linspace(-2*np.pi, 4*np.pi, 400)

# Calcular los valores de las funciones
y = np.array([f(xi) for xi in x])
n_values = [5, 10, 15]
y_sums = [sum_sin(x, n) for n in n_values]
```

```python
# Graficar todas las funciones en una sola figura
plt.figure(figsize=(8, 5))
plt.plot(x, y, label=r'$f(x)$', color='black', linewidth=2)
colors = ['b', 'g', 'r']
for y, n, c in zip(y_sums, n_values, colors):
    plt.plot(x, y, label=rf'$\sum_{{k=1}}^{{{n}}} \frac{{\sin(kx)}}{{k}}$', color=c)

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(ticks=np.arange(-2*np.pi, 4.5*np.pi, np.pi),
           labels=[r'$-2\pi$', r'$-\pi$', r'$0$', r'$\pi$', r'$2\pi$', r'$3\pi$', r'$4\pi$'])
plt.xlabel('x')
plt.ylabel('Valor de la función')
plt.title('Gráfica de f(x) y sus sumas parciales')
plt.legend()
plt.show()
```

```python
# Función de error absoluto
def error_abs(x, n):
    return np.abs(np.array([f(xi) for xi in x]) - sum_sin(x, n))

# Definir el rango de valores de x
x = np.linspace(-2*np.pi, 4*np.pi, 400)

# Valores de n para evaluar
n_values = [5, 10, 15]

# Graficar los errores absolutos
plt.figure(figsize=(8, 5))

for n in n_values:
    plt.plot(x, error_abs(x, n), label=f'n={n}')

plt.xlabel(r'$x$')
plt.ylabel(r'$|f(x) - S_n(x)|$')
plt.title('Error absoluto entre $f(x)$ y $\sum \sin(kx)/k$')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
