Dada la siguiente grafica,

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funciÃ³n a trozos
def funcion_a_trozos(x):
    if 0 <= x <= 1:
        resultado = 5/4 - x
    else:
        resultado = 1/4
    return resultado

# Crear valores de x para graficar
x_values = np.linspace(0, 2, 100)  # Rango de valores de x

# Calcular los valores de la funciÃ³n para cada valor de x
y_values = [funcion_a_trozos(x) for x in x_values]

# Graficar la funciÃ³n
plt.plot(x_values, y_values, color='blue')
plt.title('FunciÃ³n de densidad de probabilidad')
plt.xlabel('x')
plt.ylabel('f(x)')

plt.axhline(y=1/4, color='green', linestyle='--', linewidth=0.5, label='y=1/4')  # LÃ­nea horizontal en y=1/4
plt.axhline(y=5/4, color='orange', linestyle='--', linewidth=0.5, label='y=5/4')  # LÃ­nea horizontal en y=5/4

plt.grid(True)
plt.legend()
plt.show()
```

Obtenemos su funciÃ³n de densidad de probabilidad como sigue:

Observamos que se trata de una funciÃ³n a trozos, en el intervalo $[1,2]$ es la recta $y=\frac{1}{4}$, para el intervalo $[0,1]$ podemos obtener la ecuaciÃ³n usando la formula punto pendiente tomando los puntos:

$$a=\left(0,\frac{5}{4}\right)\hspace{1cm}b=\left(1,\frac{1}{4}\right)$$

Y sustituimos en la formula

$$y-y_1=\frac{y_2-y_1}{x_2-x_1}(x-x_1)$$
$$â¹y-\frac{5}{4}=\frac{\frac{1}{4}-\frac{5}{4}}{1-0}(x-0)$$
$$â¹y=\frac{5}{4}-x$$

Entonces

$$
f(x) = \left\{
            \begin{array}{ll}
	      \frac{5}{4}-x & \mathrm{si\ } 0\leq x\leq 1 \\
	      \frac{1}{4} & \mathrm{si\ } 1<x\leq 2 \\
        0 & \mathrm{otro\ caso}
            \end{array}
	      \right.\
$$

Ahora obtenemos la funciÃ³n de distribuciÃ³n acumulativa:

La funciÃ³n de distribuciÃ³n acumulativa de una variable aleatoria continua $X$ es la probabilidad de que $X$ tome un valor menor o igual a algun $x$ especifico. Esto es:

$$F(x)=\int_{-â}^x f(t)dt$$

Como la funciÃ³n esta definida a trozos, obtenemos la funciÃ³n acumulativa para cada intervalo, para el intervalo $(-â,0]$:

$$F(x)_{I(-â,0]}=\int_{-â}^x 0dt=0$$

Para el intervalo $[0,1]$:

$$F(x)_{I[0,1]}=\int_{-â}^0 0dt+\int_0^x \left(\frac{5}{4}-t\right)dt=0+\left(\frac{5}{4}t-\frac{t^2}{2}\right)\Bigg|_0^x=\frac{5}{4}x-\frac{1}{2}x^2$$

Para el intervalo $[1,2]$:

$$F(x)_{I[1,2]}=\int_{-â}^0 0dt+\int_0^1 \left(\frac{5}{4}-t\right)dt+\int_1^x \frac{1}{4}dt=0+\left(\frac{5}{4}t-\frac{t^2}{2}\right)\Bigg|_0^1+\frac{1}{4}t\Bigg|_1^x=\frac{5}{4}-\frac{1}{2}+\frac{1}{4}x-\frac{1}{4}=\frac{1}{4}x+\frac{1}{2}$$

Para el intervalo $[2,â)$:

$$F(x)_{I[2,â)}=\int_{-â}^0 0dt+\int_0^1 \left(\frac{5}{4}-t\right)dt+\int_1^2 \frac{1}{4}dt+\int_2^{-â} 0dt=0+\left(\frac{5}{4}t-\frac{t^2}{2}\right)\Bigg|_0^1+\frac{1}{4}t\Bigg|_1^2+0=\frac{5}{4}-\frac{1}{2}+\frac{1}{2}-\frac{1}{4}=1$$

Entonces,

$$
F(x) = \left\{
            \begin{array}{ll}
	      \frac{5}{4}x-\frac{1}{2}x^2 & \mathrm{si\ } 0\leq x\leq 1 \\
	      \frac{1}{4}x+\frac{1}{2} & \mathrm{si\ } 1<x\leq 2 \\
        1 & \mathrm{otro\ caso}
            \end{array}
	      \right.\
$$

Esto se muestra en la siguiente grafica.

```python
# Definir la funciÃ³n a trozos
def funcion_a_trozos(x):
    if 0 <= x <= 1:
        resultado = 5/4*x - 1/2*x**2
    elif 1 < x <= 2:
        resultado = 1/4*x+1/2
    else:
        resultado = 1
    return resultado

# Crear valores de x para graficar
x_values = np.linspace(0, 3, 100)  # Rango de valores de x

# Calcular los valores de la funciÃ³n para cada valor de x
y_values = [funcion_a_trozos(x) for x in x_values]

# Graficar la funciÃ³n
plt.plot(x_values, y_values, color='blue')
plt.title('FunciÃ³n de distribuciÃ³n acumulativa')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.grid(True)
plt.legend()
plt.show()
```

Aplicamos metodo de la transformda inversa, para esto, obtenemos la inversa de la funciÃ³n de distribuciÃ³n acumulativa.

$F^{-1}$ se define como el valor de $x$ tal que $F(x)=U$. Como la funciÃ³n es a trozos, calculamos la inversa para el intervalo $[0,1]$:

$$\frac{5}{4}x-\frac{1}{2}x^2=Uâ¹x^2-\frac{5}{2}x=-2U$$

Completamos el cuadrado del lado izquierdo para poder despejar $x$.

$$x^2-\frac{5}{2}x+\frac{25}{16}-\frac{25}{16}=-2Uâ¹\left(x-\frac{5}{4}\right)^2=-2U+\frac{25}{16}â¹x-\frac{5}{4}=\sqrt{\frac{25}{16}-2U}â¹x=\sqrt{\frac{25}{16}-2U}+\frac{5}{4}$$

Calculamos la inversa para el intervalo $[1,2]$:

$$\frac{1}{4}x+\frac{1}{2}=Uâ¹x+2=4Uâ¹x=4U-2$$

Como son las funciones inversas cambiamos los intervalos, para saber el intervalo adecuado evaluamos el limite superior del primer intervalo en su funciÃ³n acumulada y debe coincidir con la evaluaciÃ³n del limite inferior del segundo intervalo evaluado en su funciÃ³n:

$$\frac{5}{4}(1)-\frac{1}{2}(1)^2=\frac{3}{4}=\frac{1}{4}(1)+\frac{1}{2}$$

Entonces, la funciÃ³n inversa y sus intervalos son:

$$
F^{-1}(U) = \left\{
            \begin{array}{ll}
	      \sqrt{\frac{25}{16}-2U}+\frac{5}{4} & \mathrm{si\ } 0\leq U\leq \frac{3}{4} \\
	      4U-2 & \mathrm{si\ } \frac{3}{4}<U\leq 1
            \end{array}
	      \right.\
$$

Ahora generamos una muestra de nÃºmeros que cumplan esta funciÃ³n.

```python
import random as rnd
import math

def f(n): # Definir la funciÃ³n y generar n repeticiones
    lista = []
    for i in range(n):
        U = rnd.random()
        if 0 <= U <= 3/4:
            X = -((25/16)-2*U)**(1/2)+(5/4)
        else:
            X = 4*U-2
        lista.append(X)
    return lista

print(f(10))
```

```python
G = f(10000) # ObtenciÃ³n de media, varianza y mediana
media = np.mean(G)
varianza = np.var(G)
mediana = np.median(G)
print(f'Mediana: {mediana}')
print(f'Media: {media}')
print(f'Varianza: {varianza}')
```

```python
# GeneraciÃ³n del histograma de la muestra
plt.hist(G, bins=30, density=True, alpha=0.6, color='b')
plt.axvline(media, color='r', linestyle='--', label=f'Media: {media:.2f}') # Media
plt.axvline(mediana, color='g', linestyle='--', label=f'Mediana: {mediana:.2f}') # Mediana

plt.xlabel('Valor')
plt.ylabel('Densidad de probabilidad')
plt.title('Histograma')
plt.legend()

plt.tight_layout()

plt.tight_layout()
plt.show()
```

```python
# GeneraciÃ³n de funciÃ³n acumulativa de la muestra
def cdf(G):
    n = len(G)
    x = np.sort(G)
    y = np.arange(0, n)/n
    return x,y

X, Y = cdf(G)

plt.step(X, Y)
plt.axvline(media, color='r', linestyle='--', label=f'Media: {media:.2f}') # Media
plt.axvline(mediana, color='g', linestyle='--', label=f'Mediana: {mediana:.2f}') # Mediana

plt.xlabel('Valor')
plt.ylabel('Densidad de probabilidad')
plt.title('FunciÃ³n acumulativa')
plt.legend()

plt.tight_layout()

plt.tight_layout()
plt.show()
```
