En (a) y (b), se da la integral parcial $S_ùùÇ(x)$ de una funci√≥n $f$. Graficar $S_ùùÇ(x)$ para $ùùÇ=5,10,15,20,...$ luego use las graficas para determinar $f$.

(a) $S_ùùÇ(x)=\frac{1}{œÄ}[Si(ùùÇ(2-x))+Si(ùùÇ(2+x))]$

(b) $S_ùùÇ(x)=\frac{1}{œÄ}[Si(ùùÇ(1-x))+Si(ùùÇx)+Si(ùùÇ(3-x))+Si(ùùÇ(-2+x))]$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import sici
```

```python
# Definimos la funci√≥n (a)
def f(x, nu):
    Si_1, _ = sici(nu * (2 - x))
    Si_2, _ = sici(nu * (2 + x))
    return (1 / np.pi) * (Si_1 + Si_2)

# Rango de x
x = np.linspace(-5, 5, 400)

# Valores de nu
nus = np.arange(5, 105, 5)
```

```python
# Graficar
plt.figure(figsize=(10, 6))
for nu in nus:
    plt.plot(x, f(x, nu), label=f'ŒΩ={nu}')

plt.xlabel('x')
plt.ylabel('S_v(x)')
plt.title('Gr√°fica de la integral parcial para distintos valores de ŒΩ')
plt.legend(loc='best', fontsize=8, ncol=2)
plt.grid()
plt.show()
```

Del grafico anterior observamos que se presenta el fenomeno de Gibbs, esto quiere decir que tenemos puntos de discontinuidad en $(-‚àû,-2)$, $[-2,2]$ y $(2,+‚àû)$, por lo tanto la funci√≥n es:

$$
f(x)=\left\{
            \begin{array}{ll}
	      1 & \mathrm{si\ } |x|\leq 2 \\
        0 & \mathrm{otro\ caso}
            \end{array}
	      \right.\
$$

Se puede comprobar esto calculando las parciales de la integral de Fourier. Por definici√≥n:

$$
S_ùùÇ(x)=\int_0^ŒΩ [A(w)\cos(wx)+B(w)\sin(wx)] dw
$$

donde

$$
A(w)=\frac{1}{\pi}\int_{-\infty}^{+\infty} f(t)\cos(wt) dt=\frac{1}{\pi}\int_{-2}^2 \cos(wt) dt=\frac{1}{\pi}\left(\frac{2 \sin\left(2w\right)}{w}\right)
$$

y

$$
B(w)=\frac{1}{\pi}\int_{-\infty}^{+\infty} f(t)\sin(wt) dt=\frac{1}{\pi}\int_{-2}^2 \sin(wt) dt=0
$$

Entonces,

$$
S_ùùÇ(x)=\int_0^ŒΩ \frac{1}{\pi}\left(\frac{2 \sin\left(2w\right)}{w}\right)\cos(wx) dw=\frac{2}{\pi}\int_0^ŒΩ \frac{\sin\left(2w\right)}{w}\cos(wx) dw
$$

Aplicando una identidad producto a suma dentro de la integral.

$$
S_ùùÇ(x)=\frac{2}{\pi}\int_0^ŒΩ \frac{\sin\left(w(2+x)\right)+\sin\left(w(2-x)\right)}{2w} dw
$$

Haciendo $u=w(2+x)$ y $du=(2+x) dw$, $v=w(2-x)$ y $du=(2-x) dw$, se obtiene,

$$
S_ùùÇ(x)=\frac{1}{\pi}\left[\int_0^{ŒΩ(2+x)} \frac{\sin(u)}{u} du+\int_0^{ŒΩ(2-x)} \frac{\sin(v)}{v} du\right]=\frac{1}{œÄ}[Si(ùùÇ(2-x))+Si(ùùÇ(2+x))]
$$

Por lo tanto,

$$
f(x)=\left\{
            \begin{array}{ll}
	      1 & \mathrm{si\ } |x|\leq 2 \\
        0 & \mathrm{otro\ caso}
            \end{array}
	      \right.\
$$

Y graficamos esta funci√≥n

```python
# Definir la funci√≥n a trozos
def funcion_a_trozos(x):
    if -2 <= x <= 2:
        resultado = 1
    else:
        resultado = 0
    return resultado

# Crear valores de x para graficar
x_values = np.linspace(-6, 6, 100)  # Rango de valores de x

# Calcular los valores de la funci√≥n para cada valor de x
y_values = [funcion_a_trozos(x) for x in x_values]

# Graficar la funci√≥n
plt.plot(x_values, y_values, color='blue')
plt.title('Funci√≥n original')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```

```python
# Definimos la funci√≥n (b)
def g(x, nu):
    Si_1, _ = sici(nu * (1 - x))
    Si_2, _ = sici(nu * x)
    Si_3, _ = sici(nu * (3 - x))
    Si_4, _ = sici(nu * (-2 + x))
    return (1 / np.pi) * (Si_1 + Si_2 + Si_3 + Si_4)

# Rango de x
x = np.linspace(-5, 5, 400)

# Valores de nu
nus = np.arange(5, 105, 5)
```

```python
# Graficar
plt.figure(figsize=(10, 6))
for nu in nus:
    plt.plot(x, g(x, nu), label=f'ŒΩ={nu}')

plt.xlabel('x')
plt.ylabel('S_v(x)')
plt.title('Gr√°fica de la integral parcial para distintos valores de ŒΩ')
plt.legend(loc='best', fontsize=8, ncol=2)
plt.grid()
plt.show()
```

Del grafico anterior observamos que se presenta el fenomeno de Gibbs, esto quiere decir que tenemos puntos de discontinuidad en $(-‚àû,0)$, $[0,1]$, $(1,2)$, $[2,3]$ y $(3,+‚àû)$, por lo tanto la funci√≥n es:

$$
f(x)=\left\{
            \begin{array}{ll}
	      1 & \mathrm{si\ } 0 \leq x\leq 1\\
        1 & \mathrm{si\ } 2 \leq x\leq 3\\
        0 & \mathrm{otro\ caso}
            \end{array}
	      \right.\
$$

Se puede comprobar esto calculando las parciales de la integral de Fourier. Por definici√≥n:

$$
S_ùùÇ(x)=\int_0^ŒΩ [A(w)\cos(wx)+B(w)\sin(wx)] dw
$$

donde

$$
A(w)=\frac{1}{\pi}\int_{-\infty}^{+\infty} f(t)\cos(wt) dt=\frac{1}{\pi}\left[\int_{0}^1 \cos(wt) dt +\int_{2}^3 \cos(wt) dt \right]=\frac{1}{\pi}\left(\frac{\sin\left(w\right)}{w}+\frac{\sin\left(3w\right)}{w}-\frac{\sin\left(2w\right)}{w}\right)
$$

y

$$
B(w)=\frac{1}{\pi}\int_{-\infty}^{+\infty} f(t)\sin(wt) dt=\frac{1}{\pi}\left[\int_{0}^1 \sin(wt) dt +\int_{2}^3 \sin(wt) dt \right]=\frac{1}{\pi}\left(\frac{1-\cos\left(w\right)}{w}+\frac{\cos\left(2w\right)}{w}-\frac{\cos\left(3w\right)}{w}\right)
$$

Sustituyendo $A(w)$ y $B(w)$ en $S_ùùÇ(x)$ y aplicando identidades de suma/resta obtenemos

$$
S_ùùÇ(x)=\frac{1}{\pi}\int_0^ŒΩ \frac{1}{w}[\sin(wx)+\sin(w(1-x))+\sin(w(3-x))+\sin(w(-2+x))] dw
$$

Haciendo el cambio de variable adecuado para cada sumando de la integral, se obtiene,

$$
S_ùùÇ(x)=\frac{1}{\pi}\left[\int_0^{ŒΩx} \frac{\sin(u)}{u} du+\int_0^{ŒΩ(1-x)} \frac{\sin(v)}{v} dv\int_0^{ŒΩ(3-x)} \frac{\sin(c)}{c} dc+\int_0^{ŒΩ(-2+x)} \frac{\sin(t)}{t} dt\right]=\frac{1}{œÄ}[Si(ùùÇ(1-x))+Si(ùùÇx)+Si(ùùÇ(3-x))+Si(ùùÇ(-2+x))]
$$

Por lo tanto,

$$
f(x)=\left\{
            \begin{array}{ll}
	      1 & \mathrm{si\ } 0 \leq x\leq 1\\
        1 & \mathrm{si\ } 2 \leq x\leq 3\\
        0 & \mathrm{otro\ caso}
            \end{array}
	      \right.\
$$

Y graficamos esta funci√≥n

```python
# Definir la funci√≥n a trozos
def funcion_a_trozos_2(x):
    if 0 <= x <= 1:
        resultado = 1
    elif 2 <= x <= 3:
        resultado = 1
    else:
        resultado = 0
    return resultado

# Crear valores de x para graficar
x_values = np.linspace(-6, 6, 100)  # Rango de valores de x

# Calcular los valores de la funci√≥n para cada valor de x
y_values = [funcion_a_trozos_2(x) for x in x_values]

# Graficar la funci√≥n
plt.plot(x_values, y_values, color='blue')
plt.title('Funci√≥n original')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```
